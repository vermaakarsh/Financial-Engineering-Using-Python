## 2.  Fetching Data from Yahoo Finance
### Basic Stock Data Retrieval:
- Using yfinance to fetch stock data:

````python
stock = yf.Ticker('Tatamotors.NS')
````

- to fetch historical data
- Period: Duration like `'1d', '1mo', '1y',`
- Interval: Data frequency like `'1m', '5m', '1d'.`

````python
data = stock.history(period='1y')
data
````

## Basic Statistics:
Calculating mean, median, standard deviation, and other statistics.

````python
data.describe()
````

### Visualizing Stock Price Movements:
- Line Chart for historical price:

````python
plt.figure(figsize=(10,5))
plt.plot(data['Close'], label='Closing Price')
plt.title('Closing Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
````
