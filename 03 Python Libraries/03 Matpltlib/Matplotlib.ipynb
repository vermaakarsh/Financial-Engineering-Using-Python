## 2. Basic Plotting with Matplotlib
- Creating a simple line plot:


````python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4,5]
y = [10, 20, 25, 30,15]
plt.plot(x, y)
plt.show()
````

- Adding labels, titles, and legends:

````python
plt.plot(x, y, label='Line 1')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Simple Line Plot')
plt.legend()
plt.show()
````

- Customizing line styles and colors

````python
plt.plot(x, y, color='green', linestyle='--', marker='o')
plt.show()
````

The core of Matplotlib is creating simple plots, and understanding the plot() function is key to drawing 2D graphs. Customizations like labels, titles, and colors make the visualization more informative.

## 3. Advanced Plot Customizations
- Subplots: Creating multiple plots in a single figure.

````python
plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.subplot(1, 2, 2)
plt.plot(y, x)
plt.show()
````

- Changing figure size and aspect ratio:


````python
plt.figure(figsize=(10, 5))
plt.plot(x, y)
plt.show()
````

- Adding gridlines

````python
plt.plot(x, y)
plt.grid(True)
plt.show()
````

Subplots allow us to compare multiple visualizations in the same figure. Customizing figure size and adding grids enhances the readability of complex plots.

## 4. Plot Types in Matplotlib
###  Explore various plot types and their usage.
- Scatter Plot:


````python
plt.scatter(x, y)
plt.show()
````

- Bar Chart:

````python
plt.bar(x, y)
plt.show()
````

- Histogram

````python
data = [1, 1, 2, 3, 3, 3, 4, 4, 5]
plt.hist(data)
plt.show()
````

- Pie Chart:

````python
labels = ['A', 'B', 'C']
sizes = [20, 50, 30]
plt.pie(sizes, labels=labels)
plt.show()
````

Different data types require different visualization techniques. Scatter plots visualize correlations, bar charts are great for categorical data, and pie charts show proportions.



## 5. Working with Data from Pandas
 Integrate Matplotlib with Pandas for data visualization

- Plotting data directly from a Pandas DataFrame

````python
import pandas as pd
data = {'x': [1, 2, 3, 4], 'y': [10, 20, 25, 30]}
df = pd.DataFrame(data)
df.plot(x='x', y='y', kind='line')
plt.show()
````

- Customizing DataFrame plots:


````python
df.plot(x='x', y='y', kind='bar', color='red')
plt.show()
````

Matplotlib seamlessly integrates with Pandas, allowing you to plot data from DataFrames. This is especially useful for financial or stock data, where data is often stored in this format.



## 6. Advanced Matplotlib Features

Annotations: Adding text to plots for clarity.

````python
plt.plot(x, y)
plt.annotate('A Point', xy=(3, 25), xytext=(2, 30),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate('Highest Point', xy=(4, 30), xytext=(2.5, 20),arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()
````

## 9. Plotting Stock Prices with Matplotlib 

````python
import yfinance as yf
stock = yf.Ticker('TataMotors.NS')
hist = stock.history(period="1y")
plt.plot(hist.index, hist['Close'])
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title('Tata Motors Stock Price Over the Last Year')
plt.show()
````

Advanced Plot Customizations for  Comparing Stocks

````python
tickers = ['Reliance.NS', 'Tatamotors.NS', 'HdfcBank.NS']
for ticker in tickers:
    stock = yf.Ticker(ticker)
    hist = stock.history(period="1y")
    plt.plot(hist.index, hist['Close'], label=ticker)
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title('Stock Prices Comparison')
plt.legend()
plt.show()
````

## 10. Plot Types in Matplotlib : Volume and Distribution Analysis
- Bar Chart for Daily Trading Volume:

````python
plt.bar(hist.index, hist['Volume'])
plt.xlabel('Date')
plt.ylabel('Volume')
plt.title('Daily Trading Volume of Tata Motors')
plt.show()
````

- Histogram for Daily Returns Distribution:

````python
hist['Daily Return'] = hist['Close'].pct_change()
plt.hist(hist['Daily Return'].dropna(), bins=50)
plt.xlabel('Daily Return')
plt.ylabel('Frequency')
plt.title('Distribution of Tata Motors Daily Returns')
plt.show()
````

## 11. Working with Data from Pandas for Moving Averages for Stock Analysis
- Plotting Moving Averages:

````python
hist['MA50'] = hist['Close'].rolling(window=50).mean()
hist['MA200'] = hist['Close'].rolling(window=200).mean()

plt.plot(hist.index, hist['Close'], label='TATA Motors Close')
plt.plot(hist.index, hist['MA50'], label='50-Day MA')
plt.plot(hist.index, hist['MA200'], label='200-Day MA')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Tata Motors Stock Price with Moving Averages')
plt.legend()
plt.show()
````

## 12. Advanced Matplotlib Features for Annotating Events on Stock Charts

````python
earnings_dates = ['2023-08-01', '2023-11-01']  # Example earnings release dates
plt.plot(hist.index, hist['Close'])
for date in earnings_dates:
    plt.axvline(pd.to_datetime(date), color='red', linestyle='--', label='Earnings Release')
plt.title('AAPL Stock Price with Earnings Release Dates')
plt.legend()
plt.show()
````
